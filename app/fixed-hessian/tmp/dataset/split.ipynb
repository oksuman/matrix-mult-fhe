```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Read the dataset
df = pd.read_csv('12920_2018_401_MOESM1_ESM.txt')

print(f"Total samples: {len(df)}")
print(f"Features: {df.shape[1]}")
print(f"\nClass distribution:")
print(df['Cancer_status'].value_counts())

# Split 80/20
train_df, test_df = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42,
    stratify=df['Cancer_status']  # 클래스 비율 유지
)

print(f"\nTrain samples: {len(train_df)}")
print(f"Test samples: {len(test_df)}")
print(f"\nTrain class distribution:")
print(train_df['Cancer_status'].value_counts())
print(f"\nTest class distribution:")
print(test_df['Cancer_status'].value_counts())

# Save
train_df.to_csv('idash_train.csv', index=False)
test_df.to_csv('idash_test.csv', index=False)

print("\nSaved:")
print("  - idash_train.csv")
print("  - idash_test.csv")
```
